```python
url = 'https://www.althensensors.com/nl/instrumentatie/fiber-optic-instrumentatie/fiber-optic-oem-boards/5147/oem-mnp-fiber-optische-versterker/'
r = requests.get(url)
soup = BeautifulSoup(r.content)
```

```python
list_urls = []
for i in soup.find_all("ul",class_="c-nav__list js-nav__list")[0].find_all("a"):
    list_urls.append(i['href'])
```

```python
list_urls
```

```python
substring_list = ["sensoren","instrumentatie","meetsystemen","industriele-iot-oplossingen"]
```

```python
URLs = [i for i in list_urls if any(substring in i for substring in substring_list) ]
```

```python
len(URLs)
```

```python
import time
df = pd.DataFrame()
counter= 0
for link in URLs:
    print(link)
    counter= counter+1
    if counter%10== 0:
        time.sleep(10)
    r = requests.get(link)
    #print(r)
    soup = BeautifulSoup(r.content,'html.parser')
    #print(soup)
    #soup = get_soup(link)
    results = []
    hyr = soup.findAll('li',{'itemprop':'itemListElement'})
    page_hyerarchy = ''
    for i in range(0,len(hyr)):
        pg_h = hyr[i].text.strip().replace('\n','') + '|'
        page_hyerarchy = page_hyerarchy + pg_h
        #print(page_hyerarchy)
    #age_hyerarchy   
    for d in soup.findAll('div',{'class':'c-product__item'}):
        try:
            product_detail = []
            URL= d.find('a')['href']
            Product_Name= d.find('h3',{'class':'c-product__title'}).text
            Page_Header=d.find('a').find('span').find('picture').find('img')['alt'] 
            if URL is not None:
                product_detail.append(URL)
            else:
                product_detail.append(" ")
            if Product_Name is not None:
                product_detail.append(Product_Name)
            else:
                product_detail.append(" ")
            if Page_Header is not None:
                product_detail.append(Page_Header)
            else:
                product_detail.append(" ")
            product_detail.append(page_hyerarchy)
            results.append(product_detail)
            df_sample = pd.DataFrame(results,columns=['URL','Product_Name','page_Header','page_hyerarchy'])       
            #df = df.append(df_sample)
            df = pd.concat([df,df_sample.drop_duplicates()],axis=0)
        except:
            pass
        #print(d.find()
        
df = df.drop_duplicates().reset_index()
df =  df.drop(['index'],axis=1)
```

```python
df.to_csv('Finaldata.csv')
```
